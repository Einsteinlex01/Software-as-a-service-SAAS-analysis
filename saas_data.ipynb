```python
service_data.drop_duplicates(inplace = True) #drop duplicate
service_data
```

```python
service_data.groupby("country").size() #check for individual country count 
```

```python
service_data['country'] = service_data['country'].str.upper() #placing all in uppercase
```

```python
service_data
```

```python
service_data['country'] = service_data['country'].replace({'USA': 'United States', 'US': 'United States', 'UNITED STATES': 'United States', 'United states': 'United States', 'united state': 'United States', 'United States': 'United States'})

# correcting and renaming all necesarry error
```

```python
service_data
```

```python
service_data["country"] =service_data["country"].replace({"CA": "Canada", "CAN": "Canada", "CANADA": "Canada"})

# correcting and renaming all necesarry error
```

```python
service_data
```

```python
service_data["country"] =service_data["country"].replace({"UK": "United Kingdom", "UNITED KINGDOM": "United Kingdom"})


# correcting and renaming all necesarry error
```

```python
service_data["country"] =service_data["country"].replace({"NIGERIA": "Nigeria", "NGA": "Nigeria"})

# correcting and renaming all necesarry error
```

```python
service_data['country'] = service_data['country'].fillna("United States")

# correcting and renaming all necesarry error
```

```python
service_data['country']= service_data['country'].str.capitalize()

# correcting and renaming all necesarry error
```

```python
service_data
```

```python
service_data.groupby("plan_type").size() #checking plan_type details and researching about it for appropiate placement
```

```python
service_data["plan_type"] =service_data["plan_type"].fillna(method = "ffill")
```

```python
service_data["plan_type"] = service_data["plan_type"].ffill() #filling null values
```

```python
service_data
```

```python
service_data['plan_type'] = service_data['plan_type'].replace({'Basic':'Basic', 'Pro':'Premium', 'pro':'Premium', 'basic':'Basic'})

# correcting and renaming all necesarry error
```

```python
service_data.loc[service_data['monthly_revenue'] < 0, 'monthly_revenue'] = 0

#find negative values then insert 0 automatically
```

```python
plan_prices = {'Free': 0, 'Basic': 50, 'Premium': 100, 'Enterprise': 500}
service_data['monthly_revenue'] = service_data['plan_type'].map(plan_prices)

#mapping monthly_revenue with plan_type
```

```python
service_data.groupby("referral_source").size() #checking plan_type details and researching about it for appropiate placement
```

```python
service_data["referral_source"] = service_data["referral_source"].ffill() #filling forward for empty spaces
```

```python
service_data.groupby("churned").size()
```

```python
service_data['churned'] = service_data['churned'].apply(lambda x: 'No' if x == 0 else 'Yes')
print(service_data.head())
```

```python
service_data['signup_date'] = pd.to_datetime(service_data['signup_date']) #change signup_date from integer to datetime
```

```python
service_data['year'] = service_data['signup_date'].dt.year #year extraction
```

```python
service_data['month'] = service_data['signup_date'].dt.strftime('%b') #month extraction
service_data
```

```python
service_data['day'] = service_data['signup_date'].dt.day #day extraction
service_data
```
